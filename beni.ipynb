```python
# pregunta 1
import yfinance as yf
tesla = yf.Ticker("TSLA")
tesla_hist = tesla.history(period="5y")  
print(tesla_hist.head())
tesla_hist.to_csv("tesla_stock_data.csv")

```

```python
#pregunta 2
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")
tesla_revenue = []  # Lista vacÃ­a correcta

for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        rows = table.find_all("tr")
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) >= 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                if revenue != "":
                    tesla_revenue.append([date, float(revenue)])


tesla_rev_df = pd.DataFrame(tesla_revenue, columns=["Date", "Revenue"])
tesla_rev_df.head()
```
